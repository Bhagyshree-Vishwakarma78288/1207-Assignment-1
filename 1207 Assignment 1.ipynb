```python
# Importing package
import pymongo

# Connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Accessing the Database
mydb = client["database"]

# Creating Collection
collection = mydb.assignment

# Accessing Collection
collection = mydb["assignment"]
```

1.Insert the given sample data into a MongoDB collection called 'students'.

```python
# sample data 
students= [
        {"name": "Rohit", "age": 23, "math_score": 85, "physics_score": 90, "city": "New York"},
        {"name": "Eram", "age": 22, "math_score": 78, "physics_score": 75, "city": "Los Angeles"},
        {"name": "Madan", "age": 24, "math_score": 95, "physics_score": 88, "city": "Chicago"},
        {"name": "Uvaish", "age": 21, "math_score": 60, "physics_score": 65, "city": "Houston"},
        {"name": "Neha", "age": 23, "math_score": 72, "physics_score": 80, "city": "Phoenix"}
    ]

# insert many document
collection.insert_many(students)
```

discription:The output displays the details of each student in the 'students' collection, including their name, age, math score, physics score, and city. Each student record is printed as a dictionary, along with a unique _id generated by MongoDB.

2.Write a query to find all students who have a 'math_score' greater than 80.

```python
query = {"math_score": {"$gt": 80}}
for students in collection.find(query):
    print(students)
```

discription : This code retrieves and prints student records from the 'students' collection with a math_score greater than 80. It uses a query with the $gt operator to filter results, iterating through the matching documents and printing each one.

5.Write a query to find students from the city 'New York'

```python
query = {"city": "New York"}
results = collection.find(query)
for student in results:
    print(student)
```

```python
pipeline = [
    {
        "$group": {
            "_id": {
                "name": "$name",
                "age": "$age",
                "math_score": "$math_score",
                "physics_score": "$physics_score",
                "city": "$city"
            },
            "ids": {"$addToSet": "$_id"},
            "count": {"$sum": 1}
        }
    },
    {"$match": {"count": {"$gt": 1}}}
]

duplicates = list(collection.aggregate(pipeline))

# Step 2: Remove duplicates, keeping one
for duplicate in duplicates:
    collection.delete_many({"_id": {"$in": duplicate["ids"][1:]}})

print("Duplicates removed.")
```

discription : This code is collection to find all students located in "New York." It then prints each student's complete document, providing insight into their data based on the specified city filter.

9.Write a query to find students whose 'math_score' is greater than 'physics_score'.

```python
pipeline = [
    {
        "$match": {
            "$expr": { "$gt": ["$math_score", "$physics_score"] }  # Compare fields
        }
    }
]

results = collection.aggregate(pipeline)

# Print the output
for student in results:
    print(student)
```

discription:This code finds students, specifically Madan and Eram, whose math_score exceeds their physics_score using MongoDB's aggregation framework. It utilizes the $expr operator for direct field comparisons.





10.Write a query to return students sorted by 'math_score' in descending order

discription:This query retrieves all student records from the assignment collection and orders them by their math_score in descending order, ensuring that the students with the highest scores appear first. The use of sort() allows for efficient data organization based on academic performance.

11. Write a query using the aggregation framework to calculate the average 'math_score' of all students.

```python
pipeline = [
    {
        "$group": {
            "_id": None,  # Grouping all documents together
            "average_math_score": { "$avg": "$math_score" }  # Calculate average
        }
    }
]

results = collection.aggregate(pipeline)
for result in results:
    print("Average Math Score:", result["average_math_score"])
```

discription: This query calculates the average math_score of all students using the MongoDB aggregation framework. It groups all documents together with _id: None and applies the $avg operator to compute the average score is 79.5.

12.Write a query to group students by 'city' and count the number of students in each city.

```python
pipeline = [
    {
        "$group": {
            "_id": "$city",  # Group by city
            "student_count": { "$sum": 1 }  # Count the number of students in each city
        }
    }
]

# Execute the aggregation pipeline
result = collection.aggregate(pipeline)

# Print the result
for item in result:
    print(f"City: {item['_id']}, Number of Students: {item['student_count']}")
```

discription: The output of this query displays the number of students grouped by city. For each city, it shows the city name along with the total count of students in that city.
